```python
# Safe division
def divide(a, b):
    if b == 0:
        return None
    return a / b

# Examples
print(divide(10, 2))  # Output: 5.0
print(divide(10, 0))  # Output: None
```

6. (7б) Визначте функцію, яка приймає iterable (ітерабельний обʼєкт, напр. список), що складається з 7ми рядкових значень - назв погодних умов (напр. "sunny") за кожен день у тижні. Повертає функція словник, де ключами є рядки з погодними умовами, а значеннями - кількість разів, скільки дана погода була протягом тижня (напр. {'sunny': 1, ...}). Викличте цю функцію та виведіть результат її роботи на екран.

```python
# Weather counter
def count_weather(days):
    result = {}
    for day in days:
        if day in result:
            result[day] += 1
        else:
            result[day] = 1
    return result

# Example call
week = ["sunny", "rainy", "sunny", "cloudy", "sunny", "rainy", "snowy"]
print(count_weather(week))
```

число має вводитись з консолі.
власне алгоритм пошуку винесений в окрему функцію, яка повертає булеве значення (знайдено число - True, не знайдено - False).
в залежності від результату функції, виведіть на екран повідомлення про успішний або ж неуспішний пошук.
Зауваження: використовувати оператор in заборонено.

```python
# Search function
def search_number(lst, target):
    for num in lst:
        if num == target:
            return True
    return False

# Sorted list
numbers = [1, 3, 5, 7, 9, 11, 13, 15]

# User input
x = int(input("Enter number to search: "))

# Search and output
if search_number(numbers, x):
    print("Number found!")
else:
    print("Number not found.")
```

Збережіть результат у змінній, не приводяти явно до будь-якого типу. б) Використовуючи результат на кроці 12.а у якості аргументу, та за допомогою лямбда-функції в якості іншого аргументу до вбудованої функції, додайте до кожного елементу число 1 та опісля зведіть у квадрат. Остаточний результат приведіть явно до списку та виведіть на екран.

```python
# List of numbers
nums = [3, 4.5, 6, 7.0, 9.0, 10, 12.3, 15, 18.0, 21.1]

# a) Filter numbers divisible by 3
div_by_3 = filter(lambda x: x % 3 == 0, nums)

# b) Add 1 and square each number
processed = map(lambda x: (x + 1) ** 2, div_by_3)

# Convert to list and print
result = list(processed)
print("Final result:", result)
```

# III. Decorators.

11. (11б) Визначте декоратор, який би виводив у консоль повідомлення про початок роботи і назву функції (напр. my_function._ _ name _ _), а після відпрацювання функції - повідомлення про закінчення роботи і назву функції, що задекорована.

```python
# Decorator definition
def log_decorator(func):
    def wrapper(*args, **kwargs):
        print(f"Starting function: {func.__name__}")
        result = func(*args, **kwargs)
        print(f"Ending function: {func.__name__}")
        return result
    return wrapper

# Example usage
@log_decorator
def greet(name):
    print(f"Hello, {name}!")

# Call function
greet("Mark")
```

12. (4б) Задекоруйте двома різними способами функції з завдання 4 та 5 декоратором із завдання 11. (**УВАГА**: не декоруйте функції двічі!).

```python
# Decorator
def log_decorator(func):
    def wrapper(*args, **kwargs):
        print(f"Starting function: {func.__name__}")
        result = func(*args, **kwargs)
        print(f"Ending function: {func.__name__}")
        return result
    return wrapper

# Decorated with @
@log_decorator
def divide(a, b=1):
    if b == 0:
        return None
    return a / b

# Example
print(divide(10, 2))
```

15. (15б) Визначте функцію, яка рекурсивно реалізовує пошук числа у послідовності Фібоначчі за її індексом (нумерація починається з нуля). Наприклад, потрібно знайти перше число у послідовності, тоді функція прийматиме 0 у якості аргумента і повертатиме 1, якщо ж знаходити третє число (індекс 2) послідовності, то функція повертатиме 2, якщо ж знаходити десяте число (індекс 9), то функція повертатиме 55.

```python
# Recursive Fibonacci
def fibonacci(n):
    if n == 0 or n == 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

# Examples
print(fibonacci(0))   # Output: 1
print(fibonacci(2))   # Output: 2
print(fibonacci(9))   # Output: 55
```
